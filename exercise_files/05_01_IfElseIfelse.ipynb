## If statements with "FizzBuzz"

```python
# 1, 2, Fizz, 4, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, FizzBuzz, 16

for n in range(1, 101):
    if n % 15 == 0:
        print('FizzBuzz')
    else:
        if n % 3 == 0:
            print('Fizz')
        else:
            if n % 5 == 0:
                print('Buzz')
            else:
                print(n)
```

### Elif statements

```python
for n in range(1, 101):
    if n % 15 == 0:
        print('FizzBuzz')
    elif n % 3 == 0:
        print('Fizz')
    elif n % 5 == 0:
        print('Buzz')
    else:
        print(n)
```

### Single Line if statements

```python
# ternary operator - takes in some Boolean condition, evaluates it, returns one value if the condition is true,
#                    and then returns another value when the condition is false
# syntax:
# <value if true> if <Boolean condition> else <value if false>
# can assign this to a variable (see cell 11 below)

# As Python programmers, our goal is always to write clean, readable code.
# Although ternary operators easily accomplish this when used correctly, you have to be careful with them.

n = 5
print('Fizz' if n % 3 == 0 else n)
```

```python
fizzBuzz = 'Fizz' if n % 3 == 0 else n
```

```python
# ternary operators can be strung together
'Fizz' if n % 3 == 0 else 'Buzz' if n % 5 == 0 else n
```

```python
'FizzBuzz' if n % 15 == 0 else 'Fizz' if n % 3 == 0 else 'Buzz' if n % 5 == 0 else n
```

```python
# put in a list
['FizzBuzz' if n % 15 == 0 else 'Fizz' if n % 3 == 0 else 'Buzz' if n % 5 == 0 else n for n in range(1, 101)]
```
